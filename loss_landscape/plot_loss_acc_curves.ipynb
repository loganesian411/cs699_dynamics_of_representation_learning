```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read in test accuracy results
acc_test = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias-tag-test_acc.csv')
acc_test_adam = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias_adam-tag-test_acc.csv')
acc_test_5 = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias_labelnoise5-tag-test_acc.csv')
acc_test_15 = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias_labelnoise15-tag-test_acc.csv')
acc_test_25 = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias_labelnoise25-tag-test_acc.csv')
acc_test_50 = pd.read_csv('./results/acc/run-resnet20_skip_bn_bias_labelnoise50-tag-test_acc.csv')

#Read in test loss results
loss_test = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias-tag-test_loss.csv')
loss_test_adam = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_adam-tag-test_loss.csv')
loss_test_5 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise5-tag-test_loss.csv')
loss_test_15 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise15-tag-test_loss.csv')
loss_test_25 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise25-tag-test_loss.csv')
loss_test_50 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise50-tag-test_loss.csv')

# Read in train loss results
loss_train = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias-tag-train_loss.csv')
loss_train_adam = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_adam-tag-train_loss.csv')
loss_train_5 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise5-tag-train_loss.csv')
loss_train_15 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise15-tag-train_loss.csv')
loss_train_25 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise25-tag-train_loss.csv')
loss_train_50 = pd.read_csv('./results/loss/run-resnet20_skip_bn_bias_labelnoise50-tag-train_loss.csv')




# train loss is higher res --> downsample
loss_train = loss_train.iloc[::5, :]
loss_train.reset_index(drop=True, inplace=True)
loss_train_adam = loss_train_adam.iloc[::5, :]
loss_train_adam.reset_index(drop=True, inplace=True)
loss_train_5 = loss_train_5.iloc[::5, :]
loss_train_5.reset_index(drop=True, inplace=True)
loss_train_15 = loss_train_15.iloc[::5, :]
loss_train_15.reset_index(drop=True, inplace=True)
loss_train_25 = loss_train_25.iloc[::5, :]
loss_train_25.reset_index(drop=True, inplace=True)
loss_train_50 = loss_train_50.iloc[::5, :]
loss_train_50.reset_index(drop=True, inplace=True)
```

```python
fig, axs = plt.subplots(1, 3,  figsize=(16, 4))


axs[0].plot(loss_train["Value"], label="SGD + Momentum")
axs[0].plot(loss_train_adam["Value"],label="Adam")
axs[0].set_yticks(np.arange(0, 1.8, 0.25))
axs[0].legend(fontsize=12)
axs[0].set_xlabel('Epoch', fontsize=16)
axs[0].set_ylabel('Loss', fontsize=16)
axs[0].set_title('Training Loss',  fontsize=16)
axs[0].grid()
axs[0].tick_params(axis='both', which='major', labelsize=14)




axs[1].plot(loss_test["Value"], label="SGD + Momentum")
axs[1].plot(loss_test_adam["Value"], label="Adam")
axs[1].set_yticks(np.arange(0, 1.8, 0.25))
axs[1].legend(fontsize=12)
axs[1].set_xlabel('Epoch', fontsize=16)
axs[1].set_ylabel('Loss', fontsize=16)
axs[1].set_title('Test Loss',  fontsize=16)
axs[1].grid()
axs[1].tick_params(axis='both', which='major', labelsize=14)




axs[2].plot(acc_test["Value"], label="SGD + Momentum")
axs[2].plot(acc_test_adam["Value"], label="Adam")
plt.ylim(0, 1)
axs[2].legend(fontsize=12)
axs[2].set_xlabel('Epoch', fontsize=16)
axs[2].set_ylabel('Accuracy', fontsize=16)
axs[2].set_title('Test Accuracy',  fontsize=16)
axs[2].grid()
axs[2].tick_params(axis='both', which='major', labelsize=14)



fig.tight_layout(pad=1)
plt.savefig('./figures/opt.png', bbox_inches='tight')
```

```python
fig, axs = plt.subplots(1, 3,  figsize=(16, 4))
axs[0].plot(loss_train_50["Value"],label="50%")
axs[0].plot(loss_train_25["Value"],label="25%")
axs[0].plot(loss_train_15["Value"],label="15%")
axs[0].plot(loss_train_5["Value"],label="5%")
axs[0].plot(loss_train["Value"], label="0%")




axs[0].set_yticks(np.arange(0, 2.5, 0.25))
axs[0].legend(fontsize=12)
axs[0].set_xlabel('Epoch', fontsize=16)
axs[0].set_ylabel('Loss', fontsize=16)
axs[0].set_title('Training Loss',  fontsize=16)
axs[0].grid()
axs[0].tick_params(axis='both', which='major', labelsize=14)



axs[1].plot(loss_test_50["Value"], label="50%")
axs[1].plot(loss_test_25["Value"], label="25%")
axs[1].plot(loss_test_15["Value"], label="15%")
axs[1].plot(loss_test_5["Value"], label="5%")
axs[1].plot(loss_test["Value"], label="0%")




axs[1].set_yticks(np.arange(0, 2.5, 0.25))
axs[1].legend(fontsize=12)
axs[1].set_xlabel('Epoch', fontsize=16)
axs[1].set_ylabel('Loss', fontsize=16)
axs[1].set_title('Test Loss',  fontsize=16)
axs[1].grid()
axs[1].tick_params(axis='both', which='major', labelsize=14)



axs[2].plot(acc_test_50["Value"], label="50%")
axs[2].plot(acc_test_25["Value"], label="25%")
axs[2].plot(acc_test_15["Value"], label="15%")
axs[2].plot(acc_test_5["Value"], label="5%")
axs[2].plot(acc_test["Value"], label="0%")




plt.ylim(0.35, 0.95)
axs[2].legend(fontsize=12)
axs[2].set_xlabel('Epoch', fontsize=16)
axs[2].set_ylabel('Accuracy', fontsize=16)
axs[2].set_title('Test Accuracy',  fontsize=16)
axs[2].grid()
axs[2].tick_params(axis='both', which='major', labelsize=14)



fig.tight_layout(pad=1)
plt.savefig('./figures/noisylabels.png', bbox_inches='tight')

```
